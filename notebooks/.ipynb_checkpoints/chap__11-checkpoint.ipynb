```python
import numpy as np
from pylab import plt, mpl
```

```python
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
def f(x): #main function used for examples
    return np.sin(x) + 0.5 * x
```

```python
def create_plot(x, y, styles, labels, axlabels):
    plt.figure(figsize=(10, 6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    plt.legend(loc=0)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50) #x values used for the plotting and calculations
```

```python
create_plot([x], [f(x)], ['b'], ['f(x)'], ['x', 'f(x)'])
```

```python
res = np.polyfit(x, f(x), deg=1, full=True) #linear regression set up
res #full results, regression parameters, residuals, effective rank, singular values and relative condition number
```

```python
ry = np.polyval(res[0], x) #evaluation using the regression parameters
```

```python
create_plot([x,x], [f(x), ry], ['b', 'r.'],
           ['f(x)', 'regression'], ['x','f(x)'])
```

```python
#need higher order polynomials to account for sin function
reg = np.polyfit(x, f(x), deg=5)
ry = np.polyval(reg, x)
```

```python
create_plot([x,x], [f(x), ry], ['b','r.'],
           ['f(x)', 'regression'], ['x', 'f(x)'])
```

```python
reg = np.polyfit(x, f(x), deg=7)
ry = np.polyval(reg, x)
```

```python
np.allclose(f(x), ry) #checks whether the function and regression values are the same (or at least close)
```

```python
np.mean((f(x) - ry) ** 2) #calculates the mean squared error (MSE) for the regression values given the function values
```

```python
create_plot([x,x], [f(x), ry], ['b','r.'],
           ['f(x)', 'regression'], ['x', 'f(x)'])
```

### Individual Basis Functions

```python
matrix = np.zeros((3 + 1, len(x))) #the ndarray object for the basis function values (matrix)
matrix[3, :] = x ** 3 #basis functions from cubic to constant
matrix[2, :] = x ** 2
matrix[1, :] = x
matrix[0, :] = 1
```

```python
reg = np.linalg.lstsq(matrix.T, f(x), rcond=None)[0] #the regression step, central function
```

```python
reg.round(4) #the optimal regression parameters
```

```python
ry = np.dot(reg, matrix) #the regression estimates for the function values
```

```python
create_plot([x,x], [f(x), ry], ['b','r.'],
           ['f(x)', 'regression'], ['x', 'f(x)'])
```

the fit above is not as good as the polynomial fit since we used a more general approach. However, using the general approach allows to exploit knowledge about the example function, thus we must include a sine function in the set of basis functions. The highest order monomial is replaced.

```python
matrix[3, :] = np.sin(x) #the new basis function exploiting knowledge about the example function
```

```python
reg = np.linalg.lstsq(matrix.T, f(x), rcond=None)[0]
```

```python
reg.round(4)
```

```python
ry = np.dot(reg, matrix)
```

```python
np.allclose(f(x), ry) #regression now leads to a perfect fit
```

```python
np.mean((f(x) - ry) ** 2)
```

```python
create_plot([x,x], [f(x), ry], ['b','r.'],
           ['f(x)', 'regression'], ['x', 'f(x)'])
```

### Noisy Data
Regression works very well with noisy data. To illustrate this point, independent observations with noise and dependent observations with noise are generated. In a sense, the regression averages ou the noise.

```python
xn = np.linspace(-2 * np.pi, 2 * np.pi, 50) #the new deterministic x values
xn = xn + 0.15 * np.random.standard_normal(len(xn)) #introducing noise to the x values
yn = f(xn) + 0.25 * np.random.standard_normal(len(xn)) #introducing noise to the y values
```

```python
reg = np.polyfit(xn, yn, 7)
ry = np.polyval(reg, xn)
```

```python
create_plot([x,x], [f(x), ry], ['b','r.'],
           ['f(x)', 'regression'], ['x', 'f(x)'])
```
