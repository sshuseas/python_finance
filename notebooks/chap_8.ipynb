```python
import numpy as np
import pandas as pd
from pylab import mpl, plt
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
import quandl as ql
```

```python
aapl_data = ql.get('WIKI/AAPL', start_date='2010-01-01')
amzn_data = ql.get('WIKI/AMZN', start_date='2010-01-01')
msft_data = ql.get('WIKI/MSFT', start_date='2010-01-01')

eur_data = ql.get('FRED/DEXUSEU', start_date='2010-01-01')
xau_data = ql.get('WGC/GOLD_DAILY_USD', start_date='2010-01-01')
```

```python
data = pd.DataFrame(data=[aapl_data.Close, amzn_data.Close, msft_data.Close, eur_data.Value, xau_data.Value]).T
data.columns = ['aapl_close', 'amzn_close', 'msft_close', 'euro_value', 'gold_value']
data.head()
```

```python
data.plot(figsize=(15,12), subplots=True);
```

AAPL stock price dives drastically mid 2014 - Apple did a stock split to intentionally lower the price of their stock (7 new stocks at lower price for 1 previous stock)

```python
data.diff().mean()
```

```python
data.pct_change().mean().plot(kind='bar', figsize=(15,12));
```

```python
rets = np.log(data / data.shift(1)) # calculates log-returns in vectorized fashion
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(15,12));
```

```python
data = pd.DataFrame(aapl_data['Close'])
data.tail()
```

```python
window = 20 # number of index values to include for rolling window
data['min'] = data.Close.rolling(window=window).min()
data['mean'] = data.Close.rolling(window=window).mean()
data['std'] = data.Close.rolling(window=window).std()
data['median'] = data.Close.rolling(window=window).median()
data['max'] = data.Close.rolling(window=window).max()
data['ewma'] = data.Close.ewm(halflife=0.5, min_periods=window).mean() # calculates the exponentially weighted moving average, with decayomterms of half life of 0.5
```

```python
data.dropna().head()
```

```python
ax = data[['min', 'mean', 'max']].iloc[-200:].plot(figsize=(15,12), style=['g--', 'r--', 'g--'],
                                                  lw=0.8)
data.Close.iloc[-200:].plot(ax=ax, lw=2.0);
```

```python
data['SMA1'] = data.Close.rolling(window=42).mean() # sma = simple moving average
data['SMA2'] = data.Close.rolling(window=252).mean()

data[['Close', 'SMA1', 'SMA2']].plot(figsize=(15,12));
```
